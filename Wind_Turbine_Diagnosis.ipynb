# Aggreagate the data based on Year, Month, Day, Hour for better understanding of the performance of the turbine

# Calculate the no. of outliers for each column

```python
Q1 = df_new_alt.quantile(0.25)

Q3 = df_new_alt.quantile(0.75)

IQR = Q3 - Q1
IQR
```

```python
df_outliers=((df_new_alt < (Q1 - 1.5 * IQR)) | (df_new_alt > (Q3 + 1.5 * IQR))).sum()
df_outliers
```

# Data with no outliers

```python
df_no_outliers=df_new_alt[~((df_new_alt < (Q1 - 1.5 * IQR)) | (df_new_alt > (Q3 + 1.5 * IQR))).any(axis=1)]
```

# Find the feature importance of the data w.r.t. 'Active.Power'(Outliers included)
